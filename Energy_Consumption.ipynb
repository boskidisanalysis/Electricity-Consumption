```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
import plotly.express as px
```

```python
df = pd.read_csv('/content/global-energy-substitution.xls')
```

```python
df.head()
```

```python
df['Total'] = df[3:].sum(axis=0)
```

```python
df.head()
```

```python
df.Total = df.iloc[:,3:].sum(axis=1)
```

```python
df.head()
```

```python
melted = df.iloc[:,2:].melt(id_vars='Year')
melted.head()
```

```python
fig = px.area(data_frame=melted, x='Year', y='value', color='variable', labels={
    'value': 'Electricity Consumption (TWh)'
}, 
    title ='Global Electricity Consumption')

fig.for_each_trace(lambda t: t.update(name=t.name.split("=")[1]))
fig.for_each_trace(lambda t: t.update(name=t.name.split("(")[0]))
# fig.layout.yaxis.update(matches=None)

# fig.update_layout(legend=dict(
#     orientation="h",
#     yanchor="bottom",
#     y=1.02,
#     xanchor="right",
#     x=1
# ))

# fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))
# fig.for_each_annotation(lambda a: a.update(textangle=0))
# #fig.update_layout(showlegend=False)
fig.show()
```
